```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

```python
import pandas as pd
```

```python
data=pd.read_csv("/content/netflix_titles.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
sample=data[:100]
sample.description.values
```

```python
sequence=list(sample.description.values)
```

```python
sequence
```

```python
candidate_labels=["Actions","Adventure","Crime","Violence","Finance","Food"]
```

```python
output = classifier(sequence,candidate_labels,multi_label=True)
```

```python
output
```

```python
data=pd.DataFrame(output)
data
```

```python
for i in range(len(data)):
  print(len(data['scores'][i]))
```

```python
output_df=pd.DataFrame(output)
output_df
```

```python
l=[]
l1=[]
for i in range(len(output)):
  m=max(output[i]['scores'])
  a=output[i]['scores'].index(m)
  l.append(output[i]['labels'][a])
  l1.append(output[i]['scores'][a])
```

```python
output_df['label_predicted'] = l
output_df['label_scores'] =l1
```
